```python
import pandas as pd

# Sample data
data = pd.read_csv('Analysis cohorts - 2024 sin no dates (1).csv')

df = data

# Convert 'Purchase Date' to datetime with error handling
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

# Drop rows with NaT values (invalid dates)
#df = df.dropna(subset=['Purchase Date'])


df['Billed $'] = df['Billed $'].str.replace(',','')
df['Billed $'] = df['Billed $'].astype(float)
# Group by 'Connector Name', 'IN #', and 'Purchase Date' to find total billed amount per distinct order
grouped_df = df.groupby(['Connector Name', 'Purchase Date']).sum().reset_index()

# Display the result
print(grouped_df[['Connector Name', 'Purchase Date', 'Billed $']])
```
